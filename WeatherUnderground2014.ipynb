```python
import urllib2
import pandas as pd
from bs4 import BeautifulSoup

weather_data = []
date = []
#we set data sets and then we put a for loop to get the data sets under the given range
for y in range(2014,2015):
    for m in range(1, 13):
        for d in range(1, 32):
        #we check for the leap year conditionality to include in any abnormalities in the data set by function leapyear
            if y%400 == 0:
                leapyear = True
            elif y%100 == 0:
                leapyear = False
            elif y%4 == 0:
                leapyear = True
            else:
                leapyear = False

#to prevent double scraping, we check if dates have already been scraped  

            if (m==2 and leapyear and d>29):
                continue
            elif (y==2013 and m==2 and d > 28):
                continue
            elif(m in [4, 6, 9, 11] and d > 30):
                continue
#we put a timestamp on the data set
            timestamp = str(y) + str(m) + str(d)
            print 'Data on ' + timestamp

#extracting from the website's xml source and using urllib tool
            wu= 'https://www.wunderground.com/history/airport/KOAK/' + str(y) + '/' + str(m) + '/' + str(d) + '/DailyHistory.html?req_city=&req_state=&req_statename=&reqdb.zip=&reqdb.magic=&reqdb.wmo='
            weatherunderground = urllib2.urlopen(wu)

        #using soup to find the correct information from the data set
            soup = BeautifulSoup(weatherunderground)



            for row in soup.select("table tr.no-metars"):
                date.append(str(y) + '/' + str(m) + '/' + str(d))
                cells = [cell.text.strip().encode('ascii', 'ignore').decode('ascii') for cell in row.find_all('td')] #ASCI is decoded
                weather_data.append(cells)
```

```python
#We create a data frame to capture the data
weather_datadf = pd.DataFrame(weather_data)
datedf = pd.DataFrame(date)
result = pd.concat([datedf, weather_datadf], axis=1)
result
```

```python
result.to_csv('weatherdata2014.csv')   #saves it as a CSV
```
